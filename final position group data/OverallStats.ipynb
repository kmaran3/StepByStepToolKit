```python
import pandas as pd

# Ensure the team and year columns are correctly formatted for merging (if necessary)
kicking_data['team'] = kicking_data['team'].str.lower()  # Assuming team names need to be in lowercase
kicking_data['season'] = kicking_data['season'].astype(int)  # Ensure the year is an integer

AVgrades['team'] = AVgrades['team'].str.lower()  # Same assumption for AVgrades
AVgrades['season'] = AVgrades['season'].astype(int)

corrections = {
    'ind': 'clt',  # Example: changing 'ind' to 'clt'
    'ne': 'nwe',  # Example: changing 'oak' to 'rai' for Las Vegas Raiders
    'tb': 'tam',  # Example: changing 'sdg' to 'lac' for Los Angeles Chargers
    'no': 'nor',
    'la': 'ram',
    'ari': 'crd',
    'gb': 'gnb',
    'hou': 'htx',
    'lv': 'rai',
    'bal': 'rav',
    'lac': 'sdg',
    'kc': 'kan',
    'ten': 'oti',
    'sf': 'sfo'  # Example: changing 'stl' to 'lar' for Los Angeles Rams
}

kicking_data['team'] = kicking_data['team'].replace(corrections)
kicking_data = kicking_data.drop(['Unnamed: 0'], axis=1)
# Calculate points for field goals
kicking_data['FG_Points'] = (
    (kicking_data['FGM1'] + kicking_data['FGM2'] + kicking_data['FGM3']) * 3 +  # 0-39 yards all count as 3 points
    kicking_data['FGM4'] * 4 +  # 40-49 yards
    (kicking_data['FGM5'] * 5)  # 50+ yards
)

# Calculate points for missed field goals (negative points)
kicking_data['Missed_FG_Points'] = (
    -(kicking_data['FGA1'] - kicking_data['FGM1']) * 1 +  # Misses 0-19 yards
    -(kicking_data['FGA2'] - kicking_data['FGM2']) * 1 +  # Misses 20-29 yards
    -(kicking_data['FGA3'] - kicking_data['FGM3']) * 1 +  # Misses 30-39 yards
    -(kicking_data['FGA4'] - kicking_data['FGM4']) * 0.75 +  # Misses 40-49 yards
    -(kicking_data['FGA5'] - kicking_data['FGM5']) * 0.5  # Misses 50+ yards
)

# Calculate points for extra points
kicking_data['XP_Points'] = (
    kicking_data['XPM'] * 1 -  # Each PAT made
    (kicking_data['XPA'] - kicking_data['XPM']) * 0.5  # Each PAT missed
)
kicking_data['FG%'] = kicking_data['FG%'].str.replace('%', '').astype(float) / 100
kicking_data['XP%'] = kicking_data['XP%'].str.replace('%', '').astype(float) / 100
# Total fantasy score
kicking_data['fantasy_points'] = kicking_data['FG_Points'] + kicking_data['Missed_FG_Points'] + kicking_data['XP_Points']
kicking_data = kicking_data.drop(['FG_Points','Missed_FG_Points','XP_Points'], axis=1)
# kicking_data = pd.merge(kicking_data, AVgrades[['team','season','approx_value']], on=['team', 'season'], how='left')
# kicking_data['k'] = kicking_data['approx_value']
# kicking_data = kicking_data.drop(['Unnamed: 0'], axis=1)
kicking_data.to_csv("/Users/kmaran3/Dropbox/Darkhorse/final position group data/final_kicking_data.csv")
kicking_data
```
