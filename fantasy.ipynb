```python
import nfl_data_py as nfl
import pandas as pd
import numpy as np
```

```python
data_2023 = nfl.import_weekly_data([2023])
data_2023.columns 
data_2023 = data_2023[['player_id', 'player_display_name', 'position',
       'position_group','headshot_url', 'recent_team', 'season', 'week',
       'season_type', 'opponent_team', 'completions', 'attempts',
       'passing_yards', 'passing_tds', 'interceptions', 'sacks', 'sack_yards',
       'sack_fumbles', 'sack_fumbles_lost', 'passing_air_yards',
       'passing_yards_after_catch', 'passing_first_downs', 'passing_epa',
       'passing_2pt_conversions', 'pacr', 'dakota', 'carries', 'rushing_yards',
       'rushing_tds', 'rushing_fumbles', 'rushing_fumbles_lost',
       'rushing_first_downs', 'rushing_epa', 'rushing_2pt_conversions',
       'receptions', 'targets', 'receiving_yards', 'receiving_tds',
       'receiving_fumbles', 'receiving_fumbles_lost', 'receiving_air_yards',
       'receiving_yards_after_catch', 'receiving_epa',
       'receiving_2pt_conversions', 'racr', 'target_share', 'air_yards_share',
       'wopr', 'special_teams_tds', 'fantasy_points', 'fantasy_points_ppr']]
weekly_data_2023 = data_2023.sort_values(by=['player_display_name', 'week'], ascending=[True, True])
weekly_data_2023
```

```python
season_data_2023 = nfl.import_seasonal_data([2023])
grouped_data_2023 = weekly_data_2023.groupby(['player_display_name']).agg({
    'player_id': 'first'}).reset_index()
season_data_2023 = pd.merge(grouped_data_2023, season_data_2023, on='player_id', how='left')

# Display the result
print(season_data_2023)
```

```python
qb_weekly_data_2023 = weekly_data_2023[weekly_data_2023['position'] == 'QB']

numeric_cols = qb_weekly_data_2023.select_dtypes(include=['float32','int32']).columns

# Calculate the correlation between each variable and fantasy_points for quarterbacks
qb_corr_matrix = qb_weekly_data_2023[numeric_cols].corr()["fantasy_points"].abs().sort_values(ascending=False)

# Print the top 10 variables with highest correlation to fantasy_points for quarterbacks
print("Top 10 variables correlated with fantasy_points for quarterbacks:")
print(qb_corr_matrix.head(10))
```
